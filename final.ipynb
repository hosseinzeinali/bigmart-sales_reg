```python
dfnull=pd.DataFrame(data.isnull().sum().sort_values(ascending=False),columns=["null"])
dfnull["precision"]=dfnull["null"]/len(data)*100
dfnull
```

```python
data["Outlet_Size"]=data["Outlet_Size"].fillna(data["Outlet_Size"].mode()[0])
data["Item_Weight"]=data["Item_Weight"].fillna(data["Item_Weight"].mean())
dfnull=pd.DataFrame(data[["Item_Identifier","Item_Weight"]].isnull().sum(),columns=["null"])
dfnull["precision"]=dfnull["null"]/len(data)*100
dfnull
```

```python
data["Item_Fat_Content"].replace(["LF","low fat","reg"],["Low Fat","Low Fat","Regular"],inplace=True)
```

```python
data["Establishment_Year"]=data["Outlet_Establishment_Year"].apply(lambda x: 2013-x)
data=data.drop(["Outlet_Establishment_Year"],axis=1)
```

```python
X=data.drop(["Item_Identifier","Outlet_Identifier","Item_Outlet_Sales"],axis=1)
y=data["Item_Outlet_Sales"]
```

```python
X=pd.get_dummies(X)
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
```

```python
scaler.fit(X_train[n_cols])
```

```python
X_train[n_cols]=scaler.transform(X_train[n_cols])
X_test[n_cols]=scaler.transform(X_test[n_cols])
```

```python
X_train
```

```python
from sklearn.model_selection import cross_val_score
def cv(model,X_train,y_train):
    return cross_val_score(model,X_train,y_train,cv=5)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
CV=(cv(lr,X_train,y_train))
print(CV)
CVLR=CV.mean()
```

```python
from sklearn.linear_model import Lasso
lasso=Lasso()
CV=(cv(lasso,X_train,y_train))
print(CV)
CVLsoo=CV.mean()
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge()
CV=(cv(ridge,X_train,y_train))
print(CV)
CVRidge=CV.mean()
```

```python
from xgboost import XGBRegressor
regressor_xgb = XGBRegressor()
CV=(cv(regressor_xgb,X_train,y_train))
print(CV)
CVXGBreg=CV.mean()
```

```python
from catboost import CatBoostRegressor
cat=CatBoostRegressor()
CV=(cv(cat,X_train,y_train))
print(CV)
CVCatB=CV.mean()
```

```python
from xgboost import XGBRFRegressor
Xgbrf=XGBRFRegressor()
```

```python
CV=(cv(Xgbrf,X_train,y_train))
print(CV)
CVXGBRGreg=CV.mean()
```

```python
Cvscore=[CVLR,CVLsoo,CVRidge,CVXGBreg,CVCatB,CVXGBRGreg]
models=pd.DataFrame({"models":["Linearreg","Lasso","Ridge","XGBRegressor","Catboostreg","XGBRFregressor"],"Cvscore":Cvscore})
models.sort_values(by="Cvscore")
```

```python
Fin_model=XGBRFRegressor()
```

```python
params={
    'max_depth': [3, 5, 7,9 ],
    'gamma': [0.3,0.2,0.1, 0.01, 0.001],
    'subsample': [0.5, 0.7, 1],
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
FM=GridSearchCV(estimator=Fin_model,param_grid=params,cv=3,verbose=True)
```

```python
FM.fit(X_train,y_train)
```
